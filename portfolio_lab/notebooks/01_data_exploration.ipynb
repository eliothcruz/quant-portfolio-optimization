```python
import sys
from pathlib import Path

# Add project root to path when running from notebooks/
sys.path.insert(0, str(Path('..').resolve()))

import pandas as pd
import matplotlib.pyplot as plt

from src.data.loader import load_processed
from src.utils.config import load_settings, load_assets
```

```python
settings = load_settings()
tickers  = load_assets()
print(f"Tickers : {tickers}")
print(f"Period  : {settings['start_date']} → {settings['end_date']}")
```

```python
prices  = load_processed('prices_aligned.csv')
returns = load_processed('returns.csv')
prices.head()
```

```python
# Normalized price series (base 100)
normalized = (prices / prices.iloc[0]) * 100

fig, ax = plt.subplots(figsize=(12, 5))
normalized.plot(ax=ax)
ax.set_title('Normalized Price Series (base 100)')
ax.set_ylabel('Index')
ax.set_xlabel('')
plt.tight_layout()
plt.show()
```
